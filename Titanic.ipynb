# Imports and Init


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
```

```python
data = pd.read_csv('dataset\X_train.csv')
data3 = pd.read_csv('dataset\sample.csv')
data2 = pd.read_csv('dataset\X_test.csv')
data.dropna(inplace=True)
data
```

# Preprocessing


```python
data2 = data2.drop('Name', axis=1)
data2 = data2.drop('Cabin', axis=1)
data2 = data2.drop('Ticket', axis=1)
data2 = data2.drop('Fare', axis=1)
data2 = data2.drop('PassengerId', axis=1)
data2 = pd.concat([data2, pd.get_dummies(data2['Embarked'])], axis=1)
data2 = pd.concat([data2, pd.get_dummies(data2['Sex'])], axis=1)
data2 = pd.concat([data2, pd.get_dummies(data2['Pclass'])], axis=1)
data2 = data2.drop('Embarked', axis=1)
data2 = data2.drop('Sex', axis=1)
data2 = data2.drop('Pclass', axis=1)
data2.fillna(method='backfill', inplace=True)
```

```python
data = data.drop('Name', axis=1)
data = data.drop('Cabin', axis=1)
data = data.drop('Ticket', axis=1)
data = data.drop('Fare', axis=1)
data = data.drop('PassengerId', axis=1)
data = pd.concat([data, pd.get_dummies(data['Embarked'])], axis=1)
data = pd.concat([data, pd.get_dummies(data['Sex'])], axis=1)
data = pd.concat([data, pd.get_dummies(data['Pclass'])], axis=1)
data = data.drop('Embarked', axis=1)
data = data.drop('Sex', axis=1)
data = data.drop('Pclass', axis=1)
data.fillna(method='backfill', inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('Survived', axis=1), data['Survived'])
X_test.dropna(inplace=True)
y_test.dropna(inplace=True)
X_train.dropna(inplace=True)
y_train.dropna(inplace=True)
```

# Train Logistic Regression

```python
log_reg = LogisticRegression()
```

```python
grid = GridSearchCV(log_reg, {'C' : np.linspace(0.001, 1, 100)}, scoring='accuracy')
```

```python
grid.fit(data.drop('Survived', axis=1), data['Survived'])
```

```python
best = grid.best_estimator_
```

```python
best.fit(X_train,y_train)
```

# Train random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(n_estimators=20,random_state=0)
```

```python
rf.fit(X_train,y_train)
```

```python
rf_pred=rf.predict(X_test)
```

# Train SVM

```python
from sklearn.svm import SVC
```

```python
svm=SVC(kernel='sigmoid')
```

```python
svm.fit(X_train,y_train)
```

```python
svm_pred=svm.predict(X_test)
```

# Metrics


```python
accuracy_score(best.predict(X_test), y_test) #logistic regression
```

```python
metrics.mean_absolute_error(y_test,rf_pred) #random forest
```

```python
metrics.mean_squared_error(y_test,rf_pred) #random forest
```

```python
print(metrics.confusion_matrix(y_test,svm_pred)) #svm
```

```python
print(metrics.classification_report(y_test,svm_pred))
```
